# Visualize Conflict Data

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
confb = pd.read_csv('../data/Conflict Data/conflict_data_ner.csv')
confb.rename(columns={'country':'adm0_name',
                    'admin1':'adm1_name',
                    'admin2':'adm2_name',
                    'admin3':'adm3_name',
                    'actor1' : 'actor1',
                    'inter1' : 'inter1',
                    'actor2' : 'actor2',
                    'inter2' : 'inter2',
                    'interaction' : 'interaction',
                    'event_date':'date',
                    'year':'reference_year'}, inplace=True)
confb = confb[['reference_year','date','adm0_name','adm1_name','adm2_name','adm3_name','location',
        'event_type','sub_event_type','actor1','inter1','actor2','inter2','interaction','fatalities']]
confb.drop(index=0,inplace=True)
confb.reference_year = confb.reference_year.astype(str).astype(int)
confb.fatalities = confb.fatalities.astype(str).astype(int)
confb = confb[confb.reference_year.isin([2014,2015,2016,2017,2018,2019,2020])]
confb = confb[confb.adm1_name.isin(['Gao','Mopti','Tombouctou','Nord','Sahel','Est','Tahoua','Tillaberi'])]
confb = confb.reset_index(drop=True)

#confb
# Look for all different adm2_name
pd.DataFrame({'fatalities_mean' :confb.groupby(['adm0_name','adm1_name','adm2_name','adm3_name'])['fatalities'].mean()})
```

### Most Harmful Actors
For each Actor:
* total fatalities
* mean fatalities per conflict
* number of attacks


```python
# For each attacker how many fatalities per conflict
attackers = pd.DataFrame({'total_fatalities' : conf.groupby(['actor1'])['fatalities'].sum(),
                          'mean_fatalities' : conf.groupby(['actor1'])['fatalities'].mean(), 
                          'attack_numb' : conf.groupby(['actor1'])['fatalities'].count()}).sort_values(by=['total_fatalities','attack_numb','mean_fatalities'],ascending=False)
```

### Most Harmful Actors per Year

```python
# For each attackers with more than 20 total fatalities how many fatalities per conflict each year
attackers = attackers[attackers['total_fatalities'] > 20]
my_conf = conf[conf['actor1'].isin(attackers.index)]

pd.DataFrame({'total_fatalities' : my_conf.groupby(['reference_year','actor1'])['fatalities'].sum(),
              'mean_fatalities' : my_conf.groupby(['reference_year','actor1'])['fatalities'].mean(), 
              'attack_numb' : my_conf.groupby(['reference_year','actor1'])['fatalities'].count()}).sort_values(by=['total_fatalities','attack_numb','mean_fatalities'],ascending=False)
```

### Most Harmful Actors per Year in each Administration

```python
# For each attackers with more than 20 total fatalities how many fatalities per conflict each year in each adm2
pd.DataFrame({'total_fatalities' : my_conf.groupby(['reference_year','adm0_name','adm1_name','adm2_name','actor1'])['fatalities'].sum(),
              'mean_fatalities' : my_conf.groupby(['reference_year','adm0_name','adm1_name','adm2_name','actor1'])['fatalities'].mean(), 
              'attack_numb' : my_conf.groupby(['reference_year','adm0_name','adm1_name','adm2_name','actor1'])['fatalities'].count()}).sort_values(by=['total_fatalities','attack_numb','mean_fatalities'],ascending=False)
```
